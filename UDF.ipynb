# UDF (User Defined Function)

NOTE: UDF should be your last resort. Spark’s built-in functions are highly optimized; UDFs are "black boxes" to Spark and can be significantly slower.


```python
df = spark.createDataFrame([(25, "John"), (70, "Jane"), (15, "Doe")], ["age", "name"])
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

def age_group(age):
    if age < 18:
        return "minor"
    elif age< 65:
        return "adult"
    else:
        return "senior"

age_group_udf = udf(age_group, StringType())
df = df.withColumn("age_group", age_group_udf(df.age))
display(df)

```

##  This is the same results but using built in functions
