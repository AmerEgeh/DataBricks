# Explode : Unpacks arrays or maps into rows.

```python
from pyspark.sql import SparkSession

# Initialize Spark Session
spark = SparkSession.builder.getOrCreate()

# Define the data with lists (Arrays)
data = [
    (1, ["apple", "banana", "orange"]),
    (2, ["grapes", "melon"]),
    (3, [])
]

# Create DataFrame
df = spark.createDataFrame(data, ["id", "fruits"])

# Display the output
display(df)
```

```python
from pyspark.sql.functions import *
df.select('id', explode('fruits').alias('fruit')).display()
```

## posexplode

```python
from pyspark.sql.functions import *
df.select('id', posexplode('fruits').alias('index', 'fruit')).display()
```

```python
from pyspark.sql import SparkSession

# Initialize Spark Session
spark = SparkSession.builder.getOrCreate()

# Define data using dictionaries for the Map column
data = [
    (1, {"math": 90, "english": 85}),
    (2, {"math": 70, "science": 80})
]

# Create DataFrame with "id" and "scores"
df = spark.createDataFrame(data, ["id", "scores"])

# Display the output
display(df)
```

```python
from pyspark.sql.functions import *
df.select('id', explode('scores').alias('subject', 'score')).display()
```
