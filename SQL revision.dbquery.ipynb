```sql
with cte as(
select * from sales_franchises s
join
sales_transactions t on s.franchiseID=t.franchiseID
),
cte2 as (
  select sum(totalprice) total_sales, product
  from cte
  group by product
),
cte3 as (
  select *, row_number() over (order by total_sales desc) as rank
  from cte2
)

select * from cte3
where rank = :rank_num ;
```
