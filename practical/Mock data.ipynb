```python
# Notebook: 0_Setup_Mock_Data
data = [
    (1, 101, "250.00", "2023-10-01", "NEW"),
    (2, 102, "85.50", "2023-10-01", "NEW"),
    (3, 101, "-50.00", "2023-10-02", "REFUND") # Simulating a messy/negative record
]
columns = ["order_id", "customer_id", "amount", "order_date", "status"]

# Write directly as a managed Delta table into your 'mock_data' schema
spark.createDataFrame(data, columns) \
     .write \
     .mode("overwrite") \
     .saveAsTable("deve.mock_data.landing_orders")

display(spark.read.table("deve.mock_data.landing_orders"))
```
