# Diffrent Date Functions

```python
from pyspark.sql import SparkSession, functions as F

spark = SparkSession.builder.getOrCreate()

data = [
    (1, "2025-08-23"),
    (2, "2025-12-05"),
    (3, "2024-01-01"),
    (4, "2023-07-15")
]

df = spark.createDataFrame(data, ["id", "date"])

# To see the original table
display(df)
```

## date_add

```python
from pyspark.sql.functions import *
df.withColumn('date', date_add('date',30)).display()
```

## date_sub

```python
from pyspark.sql.functions import *
df.withColumn('date', date_sub('date',30)).display()
```

## date_diff

```python
from pyspark.sql.functions import *
df2 = df.withColumn('new_date', date_add('date',30))
```

```python
from pyspark.sql.functions import *
df2.withColumn('diff_days',datediff('new_date','date')).display()
```

## Add a monthe

```python
df.withColumn('date', add_months('date',5)).display()
```

## current_date

```python
from pyspark.sql.functions import *
df.select('date', current_date()).display()
```

## last_day

```python
from pyspark.sql.functions import *
df.select('date', last_day('date')).display()
```

## next_day

```python
from pyspark.sql.functions import *
df.select('date', current_date(), next_day(current_date(),'Monday')).display()
```

## year

```python
from pyspark.sql.functions import *
df.select('date', year('date')).display()
```

## month

```python
from pyspark.sql.functions import *
df.select('date', month('date')).display()
```

## quarter

```python
from pyspark.sql.functions import *
df.select('date', quarter('date')).display()
```
