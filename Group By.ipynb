# Group By

```python
data = [
    (1, "Alice", "NY", 2000),
    (2, "Bob", "CA", 1500),
    (3, "Charlie", "NY", 3000),
    (4, "David", "CA", 2500),
    (5, "Eve", "TX", 1800),
    (6, "Frank", "TX", 2200),
]

columns = ["id", "name", "state", "salary"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
display(df.groupBy('state').count())
```

```python
display(df.groupBy('state').sum('salary'))
```
