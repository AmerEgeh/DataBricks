```python
# No passwords needed here either!
bronze_path = "abfss://bronze@ameradam123.dfs.core.windows.net/sales_raw.csv"
silver_path = "abfss://silver@ameradam123.dfs.core.windows.net/sales_cleaned"

# Read the raw data
df = spark.read.csv(bronze_path, header=True, inferSchema=True)

# Clean the data
from pyspark.sql.functions import col, current_timestamp, upper
df_cleaned = df.filter(col("sales_amount").isNotNull()) \
               .withColumn("store_name", upper(col("store_name"))) \
               .withColumn("processed_at", current_timestamp())

# Save as a Delta Table
df_cleaned.write.format("delta").mode("overwrite").save(silver_path)
print("SUCCESS: Data cleaned and saved to 'silver' container.")
```
