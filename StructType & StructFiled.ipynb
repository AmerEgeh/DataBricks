# StructType  Represents the entire schema of a DataFrame (like a container for all fields).

# 

# StructField  Represents one single column/field inside that schema.

StructType is like a list of StructField objects.StructField describes name, datatype, and nullability for one column.


```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

# Schema using StructType and StructField
schema = StructType([
    StructField("id", IntegerType(), False),
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True)
])

# Each StructField("colname", DataType, Nullable) = one column.

# Sample data
data = [(1, "Alice", 30), (2, "Bob", 25), (3, None, 40)]

# Create DataFrame with schema
df = spark.createDataFrame(data, schema)
display(df)
```
