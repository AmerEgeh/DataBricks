```python
# 1. Define the paths for Silver (source) and Gold (destination)
silver_path = "abfss://silver@ameradam123.dfs.core.windows.net/sales_cleaned"
gold_path = "abfss://gold@ameradam123.dfs.core.windows.net/sales_aggregated"

# 2. Read the clean data from the Silver Delta table
df_silver = spark.read.format("delta").load(silver_path)

# 3. Create the Gold transformations (Business Aggregation)
# Let's calculate the total sales and number of transactions for each store
from pyspark.sql.functions import sum, count

df_gold = df_silver.groupBy("store_name").agg(
    sum("sales_amount").alias("total_sales_amount"),
    count("sales_amount").alias("number_of_transactions")
)

# 4. Save the aggregated data to the Gold container as a Delta table
df_gold.write.format("delta").mode("overwrite").save(gold_path)

print("SUCCESS: Data aggregated and saved to the 'gold' container.")

# 5. Display the final business-ready data!
display(spark.read.format("delta").load(gold_path))
```
