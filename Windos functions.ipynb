# Windos functions

```python
data = [
    ("Amit",   "Sales", 5000),
    ("Ravi",   "Sales", 4800),
    ("Neha",   "Sales", 4800),
    ("Suman",  "Sales", 4700),
    ("Kiran",  "Sales", 4700),
    ("Meena",  "Sales", 4700),
    ("Raj",    "Sales", 4600),
    ("Pooja",  "HR",    5200),
    ("Anil",   "HR",    5100),
    ("Ramesh", "HR",    5100),
    ("Seema",  "HR",    5000),
    ("Vikas",  "HR",    5000),
    ("Priya",  "HR",    4800),
    ("Nisha",  "HR",    4800),
    ("Arun",   "HR",    4700)
]

columns = ["name", "dept", "salary"]

df = spark.createDataFrame(data, columns)
df.display()
```

```python
df.createOrReplaceTempView('employee')
```

```python
%sql
select *, row_number() over(partition by dept order by salary desc) rn
from employee
```

ROW_NUMBER()

👉 Assigns a unique sequential number to each row within a partition (group), ordered by a column.

Even if values are the same, each row gets a different number. ✅ Example: 1,2,3,4...

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

w = Window.partitionBy('dept').orderBy(col('salary').desc())
df.withColumn('rn',row_number().over(w)).display(df)
```

RANK()

👉 Assigns a rank to rows within a partition based on order.

Ties get the same rank.

But the next rank will skip numbers. ✅ Example: 1, 2, 2, 4 (notice skip of 3).

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

w = Window.partitionBy('dept').orderBy(col('salary').desc())
df.withColumn('rn',rank().over(w)).display(df)
```

DENSE_RANK()

👉 Similar to RANK, but it does not skip numbers after ties.

Ties get the same rank.

Next rank is consecutive. ✅ Example: 1, 2, 2, 3 (no skip).

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

w = Window.partitionBy('dept').orderBy(col('salary').desc())
df.withColumn('rn',dense_rank().over(w)).display(df)
```

LAG(column, n)

👉 Fetches the value from the previous row (n steps back) in the same partition.

Useful for comparing with “previous record”. ✅ Example: Salary difference from last employee.

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

w = Window.partitionBy('dept').orderBy(col('salary').desc())
df.withColumn('rn',lag(col('salary'),1).over(w)).display(df)
```

LEAD(column, n)

👉 Fetches the value from the next row (n steps ahead) in the same partition.

Useful for comparing with “next record”. ✅ Example: Salary gap with next higher salary.

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

w = Window.partitionBy('dept').orderBy(col('salary').desc())
df.withColumn('rn',lead(col('salary'),1).over(w)).display(df)
```
