# FILTER

```python
df = spark.read.format('parquet').load('/Volumes/development/data/files/orders/orders.parquet')
display(df)
```

```python
display(df[df.status == 'Done'])
```

```python
Select_df = df.select('order_id', 'customer_name', 'status')
display(Select_df)
```

```python
filter_df = Select_df.filter(Select_df.status == 'Done')
display(filter_df)
```

```python
display(df.select('order_id','customer_name','status','payment_method'))
```

```python

display(df.select('order_id','customer_name','status','payment_method').filter((df.payment_method == 'Credit Card') & (df.status == 'Completed')))
```

```python
filter_email_df = df.filter(df.email.startswith('a'))
display(filter_email_df)
```

```python
filter_name_endswith_n_df = df.filter(df.customer_name.endswith('n'))
display(filter_name_endswith_n_df)
```

```python
display(df.filter(df.order_id < 2))
```
