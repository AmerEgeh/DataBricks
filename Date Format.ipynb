# Date Format

```python
from pyspark.sql import SparkSession, functions as F

# Initialize Spark Session
spark = SparkSession.builder.getOrCreate()

# Define the data based on the image
data = [
    (1, "2025-08-23"),
    (2, "2025-12-05"),
    (3, "2024-01-01"),
    (4, "2023-07-15")
]

# Create DataFrame
df = spark.createDataFrame(data, ["id", "date"])

# Show the original DataFrame
display(df)
```

```python
from pyspark.sql.functions import *

df.withColumn('date', date_format('date', 'yyyy-MMMM-dd')).show()
```

```python
from pyspark.sql.functions import *

df.withColumn('date', date_format('date', 'yyyy-MMM-dd')).show()
```

```python
from pyspark.sql.functions import *

df.withColumn('date', date_format('date', 'yyyy-MM-dd')).show()
```

### note that the output of date_format functino is string so you have to convert the col to date datatype

```python
from pyspark.sql.functions import *

df.withColumn('date', date_format('date', 'dd/MM/yyyy')).printSchema()
```

```python
df.withColumn('date', to_date('date', 'yyyy-mm-dd')).printSchema()
```
