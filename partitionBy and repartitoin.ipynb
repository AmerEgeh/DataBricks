# partitionBy and repartitoin in pyspark

```python
from pyspark.sql.functions import expr, row_number
from pyspark.sql.window import Window

# 1. Create a base dataframe with 11,000 rows
df_base = spark.range(1, 11001).toDF("id")

# 2. Populate the columns with random data to mimic your image
df_data = df_base.withColumn("category", expr("element_at(array('Electronics', 'Clothing', 'Home', 'Sports'), int(rand() * 4) + 1)")) \
                 .withColumn("region", expr("element_at(array('North', 'South', 'East', 'West'), int(rand() * 4) + 1)")) \
                 .withColumn("year", expr("element_at(array(2023, 2024), int(rand() * 2) + 1)")) \
                 .withColumn("month", expr("int(rand() * 12) + 1")) \
                 .withColumn("sales", expr("int(rand() * 900) + 100"))

# 3. Create the sequential 't_id' column per group using the Window functions you learned earlier!
windowSpec = Window.partitionBy("category", "region", "year", "month").orderBy("id")

# 4. Finalize the formatting to match your screenshot
df_mock = df_data.withColumn("t_id", row_number().over(windowSpec)) \
                 .select("category", "region", "year", "month", "t_id", "sales") \
                 .orderBy("category", "region", "year", "month", "t_id")

display(df_mock)
```

```python
df_mock.write.mode('overwrite').partitionBy('region').parquet("/Volumes/development/data/files/ales")
```
