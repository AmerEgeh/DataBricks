# Union

```python
data1 = [
    (1, "Alice", "HR"),
    (2, "Bob", "IT")
]
columns = ["id", "name", "dept"]

df1 = spark.createDataFrame(data1, columns)
display(df1)
```

```python
data2 = [
    (3, "Charlie", "Finance"),
    (4, "David", "IT"),
    (2, "Bob", "IT")   # duplicate row
]

df2 = spark.createDataFrame(data2, columns)
display(df2)
```

```python
display(df1.union(df2).distinct())
```
